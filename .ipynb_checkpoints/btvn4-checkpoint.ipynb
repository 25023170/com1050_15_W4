```python
#bài 2
def kiem_tra_so_nguyen_to():
    while True:
        n = int(input("Nhập vào một số nguyên dương: "))
        if n > 0:              
            break
        print("Số vừa nhập không phải số nguyên dương, vui lòng nhập lại!")
    if n < 2:
        print(f"{n} không phải là số nguyên tố.")
        return
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            print(f"{n} không phải là số nguyên tố.")
            return
    print(f"{n} là số nguyên tố.")   
kiem_tra_so_nguyen_to()
```

```python
#bai4
def dem_chu_so(n):
    n = abs(n)
    if n == 0:
        return 1
    count = 0
    while n > 0:
        count += 1
        n //= 10
    return count
n = int(input("Nhập số: "))
print("Nhập số: ", dem_chu_so(n))
```

```python
#bai 8
def lay_so_dao_nguoc(k):
    return int(str(k)[::-1])
def la_so_doi_xung(k):
    return k == lay_so_dao_nguoc(k)
n = int(input("Nhập số nguyên dương n: "))

step_count = 0 
while not la_so_doi_xung(n):
    rev_n = lay_so_dao_nguoc(n) 
    n = n + rev_n               
    step_count += 1      
print("-------------------------")
print(f"Tổng số bước: {step_count}")
print(f"Số đối xứng thu được: {n}")
```
